#### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Emirates_reviews_data_with_sentiment.csv')
df.head()
```

```python
df['Sentiment'].value_counts()
```

```python
df['Sentiment'] = df['Sentiment'].replace('Neutral', 'Negative')
```

```python
df['Sentiment'].value_counts()
```

#### Will use only Negative and Positive Reviews

```python
#data = df[df['Sentiment'] != 'Neutral'].copy()
#data.head()
```

#### Converting the Categorical Target Variable(Sentiment) into Numerical.
#### Will use Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df['Sentiment'] = le.fit_transform(df['Sentiment'])
df['Sentiment'].unique()
```

```python
df.head()
```

#### 0 -> Negative  |      1 -> Positive  

```python
df['Sentiment'].value_counts()
```

### TF-IDF Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['corpus']).toarray()
```

```python
X = pd.DataFrame(X)
X.head()
```

```python
X.shape
```

```python
y = df['Sentiment']
y.head()
```

```python
y.shape
```

### Splitting the data into training and test set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
print(f"X train: {X_train.shape}")
print(f"y train: {y_train.shape}")
print(f"X test: {X_test.shape}")
print(f"y test: {y_test.shape}")
```

### Model Building

```python
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score, roc_auc_score, classification_report,confusion_matrix, ConfusionMatrixDisplay

def evaluate_model(y_true, y_pred, model_name="Model", average='macro'):
    
    # Calculating metrics
    accuracy = round(accuracy_score(y_true, y_pred) * 100, 2)
    f1 = round(f1_score(y_true, y_pred, average=average,zero_division=0) * 100, 2)
    recall = round(recall_score(y_true, y_pred, average=average,zero_division=0) * 100, 2)
    precision = round(precision_score(y_true, y_pred, average=average,zero_division=0) * 100, 2)
    #roc_auc = round(roc_auc_score(y_true, y_pred, average=average, multi_class='ovr') * 100, 2)

    # Printing results
    print(f"{model_name} Evaluation Metrics:")
    print(f"Accuracy: {accuracy}%")
    print(f"F1 Score: {f1}%")
    print(f"Recall: {recall}%")
    print(f"Precision: {precision}%")
    #print(f"ROC AUC Score: {roc_auc}%")
    print("\nClassification Report:")
    print(classification_report(y_true, y_pred, zero_division=0))

    return {
        "accuracy": accuracy,
        "f1_score": f1,
        "recall": recall,
        "precision": precision,
        #"roc_auc_score": roc_auc
    }
```

```python
model_score = {}
```

```python
def plot_confusion_matrix(cm,title='Confusion Matrix',cmap=plt.cm.Greens,display_metrics=True):
    plt.figure(figsize=(7, 5))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()

    classes = ['Negative','Positive'] # 0 ->Negative, 1->Positive
    plt.xticks(ticks=np.arange(len(classes)), labels=classes)
    plt.yticks(ticks=np.arange(len(classes)), labels=classes)

     # Add text annotations to each cell
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j, i, f'{cm[i, j]}',
                     ha='center', va='center',
                     color='white' if cm[i, j] > cm.max() / 2. else 'black')
            
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.grid(False)
```

### 1. Multinomial Naive Baiyes

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
mnb = MultinomialNB()
mnb.fit(X_train, y_train)
```

```python
# Make predictions
y_pred_mnb = mnb.predict(X_test)

metrics = evaluate_model(y_test, y_pred_mnb, model_name="MNB with TF-IDF")

model_score["MNB with TF-IDF"] = list(metrics.values())
```

```python
mnb_cm = confusion_matrix(y_test, y_pred_mnb)
plot_confusion_matrix(mnb_cm)
```

#### The model performs well in identifying Positive instances with high recall and decent precision,
#### but it struggles with identifying Negative instances, resulting in very low recall for the Negative class. 
#### This imbalance suggests that while the model is effective for the Positive class, it needs improvement for handling the Negative class effectively.

### 2. SVM

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

```python
svm = SVC()

param_grid = {
    'C': [0.1, 1, 10, 100],           # Regularization parameter
    'kernel': ['linear', 'rbf', 'poly'],  # Kernel types
    'gamma': ['scale', 'auto', 0.001, 0.01, 0.1, 1]  # Kernel coefficient
}

grid_search = GridSearchCV(
    estimator=svm,             # SVM model
    param_grid=param_grid,     # Parameter grid
    cv=5,                      # Number of cross-validation folds
    scoring='f1_weighted',     # Scoring metric 
    n_jobs=-1,                 # Use all available cores
    verbose=2                  # Verbosity level 
)

grid_search.fit(X_train, y_train)
```

```python
best_svm = grid_search.best_estimator_
print(best_svm)

y_pred_svm = best_svm.predict(X_test)

metrics = evaluate_model(y_test, y_pred_svm, model_name="SVM")

model_score["SVM"] = list(metrics.values())
```

```python
svm_cm = confusion_matrix(y_test, y_pred_svm)
plot_confusion_matrix(svm_cm)
```

#### The SVC model performs well overall with higher precision, recall, and F1-score for the Positive class compared to the Negative class. 
#### The model is effective at identifying Positive instances, while its performance for Negative instances is somewhat lower but still reasonable. 
#### The overall accuracy of 81.43% and F1 score of 78.13% suggest a strong performance in balancing the classification tasks.

### 3. Tree Based Models

#### 3.1 Decision Tree Classifier

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
```

```python
# Make predictions
y_pred_dtc = dtc.predict(X_test)

metrics = evaluate_model(y_test, y_pred_dtc, model_name="Decision Tree Classifier")

model_score["Decision Tree Classifier"] = list(metrics.values())
```

#### 3.2 Using Grid Search CV

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'criterion': ['gini', 'entropy'],       # Criteria to split
    'max_depth': [None, 10, 20, 30, 40, 50], # Maximum depth of the tree
    'min_samples_split': [2, 5, 10], # Minimum number of samples required to split an internal node
    'min_samples_leaf': [1, 2, 4],    # Minimum number of samples required to be at a leaf node
    'max_features': [None, 'sqrt', 'log2'] # Number of features to consider when looking for the best split
}

dtc = DecisionTreeClassifier()

grid_search = GridSearchCV(estimator=dtc, param_grid=param_grid, 
                           cv=5, n_jobs=-1, verbose=2, scoring='accuracy')

grid_search.fit(X_train, y_train)
```

```python
# Get the best estimator
best_dtc = grid_search.best_estimator_
print(best_dtc)

# Make predictions using the best estimator
y_pred_dtc = best_dtc.predict(X_test)
```

```python
metrics = evaluate_model(y_test, y_pred_dtc, 
                         model_name="Decision Tree Classifier with GridSearchCV")

model_score["Decision Tree Classifier with GridSearchCV"] = list(metrics.values())
```

```python
dtc_GSCV_cm = confusion_matrix(y_test, y_pred_dtc)
plot_confusion_matrix(dtc_GSCV_cm)
```

#### 4.3 XG Boosting

```python
from xgboost import XGBClassifier
```

```python
xgb_clf = XGBClassifier(
    n_estimators=100,       # Number of boosting rounds
    learning_rate=0.1,      # Step size for each boosting round
    max_depth=3,            # Maximum depth of the trees
    random_state=42         # Ensures reproducibility
)

# Fit the model
xgb_clf.fit(X_train, y_train)
```

```python
# Make predictions
y_pred_xgb = xgb_clf.predict(X_test)

metrics = evaluate_model(y_test, y_pred_xgb, 
                         model_name="XG Boosting")

model_score["XGBoost"] = list(metrics.values())
```

```python
xgb_cm = confusion_matrix(y_test, y_pred_xgb)
plot_confusion_matrix(xgb_cm)
```

```python
param_grid = {
    'n_estimators': [50, 100, 150],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5]
}

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=XGBClassifier(random_state=42), 
                           param_grid=param_grid,
                           cv=5, scoring='f1', verbose=2, n_jobs=-1)

# Fit GridSearchCV
grid_search.fit(X_train, y_train)
```

```python
# Best parameters and best score
print("Best Parameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)

# Evaluate the best model
best_xgb = grid_search.best_estimator_

y_pred_xgb = best_xgb.predict(X_test)

metrics = evaluate_model(y_test, y_pred_xgb, model_name="XG Boosting GridSearch")

model_score["XGBoost with GridSearch"] = list(metrics.values())
```

```python
xgbGridSearch_cm = confusion_matrix(y_test, y_pred_xgb)
plot_confusion_matrix(xgbGridSearch_cm)
```
